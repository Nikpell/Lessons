Задание 2.
Выполните разметку на основе правил (rule-based labeling) на подмножестве выбранного датасета. Разработайте и реализуйте набор правил или условий, которые позволят автоматически присваивать метки данным на основе определенных шаблонов или критериев.

```python

from textblob import TextBlob
pd.options.mode.copy_on_write = True
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity


data['sentiment'] = data['review'].apply(get_sentiment)
data.describe()
```

```python
def get_sentiment_label(score):
    if score > 0:
        return 'Positive'
    return 'Negative'
data['sentiment_label'] = data['sentiment'].apply(get_sentiment_label)
data.head()
```

Задача 3.
Выполните разметку вручную отдельного подмножества выбранного датасета с помощью выбранного вами инструмента разметки.

```python
subset_handle = data.sample(frac=0.01)
subset_handle.to_csv('subset.csv', index=False)
path_sudset = '/home/nic/Downloads/project-6-at-2024-06-02-12-05-d3280140.csv'
subset_handle = pd.read_csv(path_sudset)

subset_handle.head()
```

Задача 4.
Объедините данные, размеченные вручную, с данными, размеченными на основе правил. Объедините два подмножества размеченных данных в один набор данных, сохранив при этом соответствующую структуру и целостность.


```python
handle = subset_handle[['Unnamed: 0', 'condition', 'drugName', 'review', 'rating', 'sentiment']]
handle.rename(columns = {'Unnamed: 0':'', 'sentiment':'sentiment_handel'}, inplace = True )
handle['sentiment'] = handle['review'].apply(get_sentiment)
handle = handle[['','drugName', 'condition', 'review', 'rating', 'sentiment', 'sentiment_handel']]
handle.head(24)
```

```python
handle = handle.set_index('')
merged = pd.merge(data, handle, how='left')
merged.head()
```

```python
merged['sentiment_handel'] = merged['sentiment_handel'].fillna(merged['sentiment_label'])
smth = merged[merged['sentiment_handel'] != merged['sentiment_label']]
smth.head()
```

Задача 5.
Обучите модель машинного обучения, используя объединенный набор размеченных данных. Разделите датасет на обучающий и тестовый наборы и используйте обучающий набор для обучения модели.

```python
from sklearn.model_selection import train_test_split # Для разделения данных на обучающую и тестовую выборку
from sklearn.feature_extraction.text import TfidfVectorizer # для преобразования текста в вектор
from sklearn.linear_model import LogisticRegression # использование модели логистической регрессии
from sklearn.metrics import accuracy_score, classification_report # оценка производительности модели
from sklearn.pipeline import Pipeline # конвеер обработки данных


X_train, X_test, y_train, y_test = train_test_split(
merged['review'], # текст рецензии - будет использоваться как входные данные
merged['sentiment_handel'], # метка классов (положительный или отрицательный )
test_size=0.2, # доля данных которая попадет в тест (20%)
random_state=42 # Зерно генератора случайных чисел для воспроизводимости результатов
)
pipeline = Pipeline([
('tfidf', TfidfVectorizer(stop_words='english')), # векторизация текста, с исключением стоп слов англ языка
('clf', LogisticRegression(max_iter=1000)) # модель логистической регресии с увелечинным колличеством итераций до 1000
])
pipeline.fit(X_train, y_train)
```

Задача 6.
Оценить эффективность обученной модели на тестовом датасете. Используйте подходящие метрики оценки. Интерпретируйте результаты и проанализируйте эффективность модели в решении задачи разметки.

```python
predictions = pipeline.predict(X_test) # предсказание модели на тестовых данных
accuracy = accuracy_score(y_test, predictions) # расчет точности модели
report = classification_report(y_test, predictions) # оценка качесвта предсказания

print(f"Точность модели {accuracy}")
print('Отчет по классификации')
print(report)
```

Данная модель имеет высокую точность предсказания. Однако низкий процент истинно положительных предсказаний для негативныых оценок низок, что не дает возможности рекомендовать ее в качестве эффективной модели.
